```python
%%ai vioras
楕円の短軸の頂点を作図する際に、焦点からの距離が長軸半径となることから、長軸との交点に焦点は自然に現れます
```

```python
%%ai vioras
こうして得られた焦点を中心とする長軸半径の円、または長軸半径の二倍の円と焦点を基点とする補助線によって
焦点からの距離の和が長軸径に等しいことを示すことができます。
```
