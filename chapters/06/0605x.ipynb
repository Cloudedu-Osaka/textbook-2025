```python
%%ai vioras
交わらない二つの円の共通接線は、円柱や円錐の切断面に現れる楕円と密接に関連しています。
```

```python
%%ai vioras
半径が等しい交わらない円の共通接線は、内部共通接線の交点において対称です。
```

```python
%%ai vioras
すなわち、円柱の切断面に現れる楕円の長軸は、中心において対象です。
```

```python
%%ai vioras
半径が異なる交わらない二つの円の共通接線は、内部接線の交点において対称性が失われています。
```

```python
%%ai vioras
このため、対応する円錐の切断面に現れる「楕円」において、楕円の中心は円錐の回転軸上にありません。この性質が円錐曲線の研究を難解にしています。
```

```python
%%ai vioras
二つの円 (円錐の場合はDandelin球面) の中心の中点を取ります。それを中心とする円のうち、接線の交点を点を通る円は、他のすべての交点と交わります。
この円の「楕円」の中心を通る直径における対称性から、楕円の長軸の対称性を証明することができます。
```
