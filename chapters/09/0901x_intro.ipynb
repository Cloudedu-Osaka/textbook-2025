```python
%%ai vioras
Newtonは、円錐曲線が二次多項式であり、幾何平均によって解析できることを知っていました。
```

```python
%%ai vioras
Newtonは、二次多項式で得た知見が、高次元多項式、さらに一般化された二項定理によって三角関数にも応用できることを発見し、微分を定義しました。
```

```python
%%ai vioras
Newtonに先立って、Fermatは、二変数の和の冪乗に関する展開公式から、多項式の接線の公式を発見しました。
```
