## 8.1 KeplerとGalileo

* [Kepler's equation - Wikipedia](https://en.wikipedia.org/wiki/Kepler%27s_equation#Inverse_problem)#Inverse_problem
    * "I am sufficiently satisfied that it [Kepler's equation] cannot be solved a priori, on account of the different nature of the arc and the sine. But if I am mistaken, and any one shall point out the way to me, he will be in my eyes the great Apollonius."

::::{admonition} ask ai
:class: tip

* "a priori" とは?

:::{admonition} ask ai repeatedly
:class: seealso

* KeplerはKepler方程式を "a priori" に解けないと言いました
* Keplerの科学的な仮説は的中します。同時代のGalileoと比べても科学的洞察の深さが違います
* KeplerとGalileoを比較し、Galileoの轍を踏まず、Keplerのようにあるべき道を示してください
* Keplerですら、ハレー彗星を観測したにもかかわらず、それが楕円軌道であることを洞察できませんでした
* GalileoやKeperの時代、Apolloniusは学術的権威でした
* Apolloniusに先立ったMenaechmusは、放物線の高さが幅の二乗に比例することを知っていました。
  また、楕円の幅と高さの関係も知っていました
* Apolloniusの円錐曲線論は、Menaechmusの発見を尊重していません。
  ルネサンスの科学者が正解に近づくには、Apolloniusの権威に縋るのではなく、
  自分でMenaechmusの成果を再発見する必要がありました
* Galileoは、放物運動の軌跡の本質に迫ったにもかかわらず、それが放物線だと見抜くことができませんでした
